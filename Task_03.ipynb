```python
import pandas as pd
data=pd.read_csv('student_feedback.csv')
data.head(1002)
```

```python
data.count(0,1)
```

```python
data.columns
```

```python
data = data.drop(columns=['','column1'], errors='ignore')
data.columns=[col.strip().replace(" ","_").lower() for col in data.columns]
display(data)
```

```python
data.info()
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

graph=data.drop(columns='student_id')
ratings=graph.mean()

plt.figure(figsize=(10,5))
sns.scatterplot(y=ratings.index,x=ratings.values)
plt.xlabel('Feedback Category')
plt.ylabel('Average Rating')
plt.title('Average Ratings for Different Feedback Categories')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

graph=data.drop(columns='student_id')
ratings=graph.min()
ratings2=graph.max()

plt.figure(figsize=(10,5))
sns.pointplot(y=ratings.index,x=ratings.values)

sns.pointplot(y=ratings2.index,x=ratings2.values)
plt.xlabel('Feedback Category')
plt.ylabel('Minimum And Maximum Rating')
plt.title('Minimum And Maximum Ratings for Different Feedback Categories')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

graph=data.drop(columns='student_id')
ratings=graph.mean()

plt.figure(figsize=(10,5))
sns.barplot(y=ratings.index,x=ratings.values)
plt.xlabel('Feedback Category')
plt.ylabel('Average Rating')
plt.title('Average Ratings for Different Feedback Categories')
plt.show()
```
